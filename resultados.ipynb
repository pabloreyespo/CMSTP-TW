```python
import os
import pandas as pd
```

```python
FOLDER = "irace_experiments"
DIR = os.listdir(FOLDER)
```

```python
dfs = []
for excel in DIR:
    name, Q, n = excel.rstrip(".xlsx").split(" ")
    Q = Q[2:]
    n = n[2:]
    df = pd.read_excel(f"{FOLDER}/{excel}")
    df["nombre_experimento"] = name
    df["Q"] = int(Q)
    df["n"] = int(n)
    df["codigo"] = df["name"] +"-"+ n +"-"+ Q
    dfs.append(df)
df = pd.concat(dfs)
# df.index = ["Q", "n"]
```

```python
bks = df[["min","codigo"]].groupby("codigo").min()
bks = bks.reset_index()
bks.columns = ["codigo","bks"]
df_bks = df.merge(bks, how = "left", on ="codigo")
df_bks["gap_min"] = (df_bks["min"]- df_bks["bks"]) / df_bks["bks"]
df_bks["gap_avg"] = (df_bks["avg"]- df_bks["bks"]) / df_bks["bks"]
df_bks
```
